Jupyter Notebooks use two primary types of cells: Code Cells and Markdown Cells. ## 1. Data Analysis and Visualization:

Python provides powerful libraries for both data analysis and visualization, making it an ideal tool for empirical research. Among these, Matplotlib and Seaborn are two essential libraries for creating high-quality visualizations. Matplotlib is a versatile library that allows users to create a wide range of static, animated, and interactive plots, making it the foundation of Python’s visualization ecosystem. On top of that, Seaborn builds on Matplotlib, offering a more intuitive interface and beautiful default styles for statistical visualizations, such as histograms, box plots, and regression plots.

````python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style for better aesthetics
sns.set_style("darkgrid")

# Generate a sample dataset
np.random.seed(42)  # For reproducibility
data = pd.DataFrame({
    "x": np.linspace(1, 100, 100),  # A simple range
    "y": np.linspace(1, 100, 100) + np.random.normal(0, 10, 100),  # Adding some noise
    "category": np.random.choice(["A", "B", "C"], 100)  # Categorical variable
})
# Display the first few rows of the dataset
#print(data.head())
data.head()
````

````python

## ---- Matplotlib Example: Basic Line Plot ---- ##
plt.figure(figsize=(8,5))
plt.plot(data["x"], data["y"], color="blue", linestyle="--", marker="o", label="Noisy Linear Data")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Matplotlib: Basic Line Plot Example")
plt.legend()
plt.show()

## ---- Seaborn Example: Scatter Plot with Regression Line ---- ##
plt.figure(figsize=(8,5))
sns.regplot(x="x", y="y", data=data, scatter_kws={"color": "darkblue"}, line_kws={"color": "red"})
plt.title("Seaborn: Regression Line with Scatter Plot")
plt.show()

## ---- Seaborn Example: Box Plot ---- ##
plt.figure(figsize=(8,5))
sns.boxplot(x="category", y="y", data=data, palette="coolwarm")
plt.title("Seaborn: Box Plot of Y by Category")
plt.show()

## ---- Seaborn Example: Distribution Plot ---- ##
plt.figure(figsize=(8,5))
sns.histplot(data["y"], kde=True, bins=20, color="green")
plt.title("Seaborn: Distribution of Y with KDE")
plt.xlabel("Y values")
plt.show()
````

## 2. Statistical and Econometric Modeling 

Statistical and econometric modeling relies on random numbers, probability distributions, and probability calculations to analyze real-world data and test economic theories. In empirical research, we often work with random variables that follow specific probability distributions, such as the normal distribution (e.g., modeling stock returns), the Poisson distribution (e.g., modeling the number of customer arrivals), or the binomial distribution (e.g., modeling success/failure outcomes). These distributions help us simulate real-world scenarios, generate synthetic data, and quantify uncertainty. Using Python, we can draw random samples from these distributions and compute key probabilities. For instance, we can answer questions like: What is the probability that a stock return is below a certain threshold? or How likely is it that a firm will experience a given number of customer visits in an hour? Python's numpy and scipy.stats libraries provide powerful tools to generate random numbers, calculate probabilities, and estimate statistical properties, making it an essential toolkit for econometric modeling and hypothesis testing. Let’s explore some examples in Python!

````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
````

````python
# 1. Generate random numbers from a normal distribution (mean=0, std=1)
normal_data = np.random.normal(loc=0, scale=1, size=1000)
````

````python
# 2a. Probability of a value being <= 1.5 in a standard normal distribution
prob_a = stats.norm.cdf(1.5)  
print(f"Probability P(X <= 1.5) in N(0,1): {prob_a}")

# 2b. Probability of a value being >= 2 in a standard normal distribution
prob_b = 1 - stats.norm.cdf(2) 
print(f"Probability P(X >= 2) in N(0,1): {prob_b}")

# 2c. Probability of a value being <= 1.5 but also >= 2 in a standard normal distribution
prob_c = stats.norm.cdf(2) - stats.norm.cdf(1.5) 
print(f"Probability P(1.5 <= X <= 2) in N(0,1): {prob_c}")
````

````python
# 3. Visualizing probability distributions
x = np.linspace(-4, 4, 1000)
plt.plot(x, stats.norm.pdf(x), label="Normal Distribution (μ=0, σ=1)", color='blue')
plt.xlabel("X values")
plt.ylabel("Probability Density")
plt.title("Probability Density Function of a Normal Distribution")
plt.legend()
plt.show()
````

We want to fit the model: 

$$ satisfaction_i = \beta_0 + \beta_1 sports_i + u_i$$


````python
# Define dependent and independent variables
X = sm.add_constant(df['sports']) 
y = df['satisfaction']

# Fit the model
slr_model = sm.OLS(y, X).fit()
print(slr_model.summary())
````
